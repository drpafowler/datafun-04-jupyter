```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.ion()
df = sns.load_dataset('penguins')
```

```python
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# Inspect histogram by numerical column
df['body_mass_g'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

```python
# Renaming a column
df.rename(columns={'species': 'Penguin Species'}, inplace=True)

# Adding a new column in lbs
df['body_mass_lbs'] = df['body_mass_g'] * 0.00220462

# Adding a new column in kg
df['body_mass_kg'] = df['body_mass_g'] / 1000

print(df.head())
```

```python
# This cell is Python too - you'll need to Run the previous cell first

# Create a scatter plot of penguin flipper length vs. body mass - use the variable name that holds your DataFrame
plt = sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="Penguin Species")

# Set axis labels and chart title
plt.set_xlabel("Flipper Length (mm)")
plt.set_ylabel("Body Mass (g)")
plt.set_title("Penguin Flipper Length vs. Body Mass")

# Run this cell by clicking the play button (Execute cell and below) in the cell toolbar
# Or by pressing Shift+Enter 
# or by clicking Run All in the Menu above
```

```python
# Create a scatterplot of bill length vs bill depth
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=df, hue="Penguin Species")
plt.set_xlabel("Bill Length (mm)")
plt.set_ylabel("Bill Depth (mm)")
plt.set_title("Bill Length vs Bill Depth")
```

```python
# Create a histogram of flipper length
sns.histplot(x = "flipper_length_mm", data = df, hue = "Penguin Species", kde=True)
plt.set_title ("Flipper Length (mm)")
```

```python
# Are penguins sexually dimorphic? Create a boxplot of flipper length by sex and species
sns.boxplot(x = "Penguin Species", y = "flipper_length_mm", data =df, hue = "sex")
plt.set_title("Flipper Length for 3 Penguin Species by Sex")
 
```
